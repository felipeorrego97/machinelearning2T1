```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
my_face = cv2.imread('/content/AndresFelipeOrrego.jpg')
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(my_face, cmap='gray')
plt.axis('off')
plt.show()
```

```python
class_faces = []
path = '/content/photos'
```

```python
for filename in os.listdir(path):
    # Check if the file is an image (you can add more image extensions if needed)
    if filename.endswith(('.jpg', '.jpeg', '.png')):
        # Construct the full file path
        file_path = os.path.join(path, filename)

        # Load the image in grayscale
        image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)

        # Resize the image to 256x256 pixels
        image_resized = cv2.resize(image, (256, 256))

        # Append the resized image to the list
        class_faces.append(image_resized)





```

```python
average_face = np.mean(class_faces, axis=0)
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(average_face, cmap='gray')
plt.title("Average Face")
plt.axis('off')
plt.show()
```

```python
distancia = np.linalg.norm(my_face - average_face)
```
